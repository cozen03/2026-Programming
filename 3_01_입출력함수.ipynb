#변환함수
int(): 정수로 변환

float(): 실수로 변환

str(): 문자로 변환

````python
a = input("이름: ")
b = input("나이: ")
print(a , "님 안녕하세요")
print(20 - int(b) , "년 뒤에 20살이네요")
````

````python
a = input("가격(원): ")
b = input("할인율(%): ")
c = int(a) * int(b) / 100
print(f"할인된 가격은 {c:0.0f}입니다")
````

````python
#섭씨 (℃)온도를 화씨 온도(℉)로 변환하는 프로그램

a = float(input("섭씨 온도를 입력하세요: "))

b = (9 * a / 5) + 32

print(f"섭씨 {a}도는 화씨 {b}도 입니다.")
````
