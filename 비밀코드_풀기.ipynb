```python
#빈리스트 만들기
animals = []
#5개의 요소로 이루어진 리스트 만들기
animals = ['개', '개미', '고양이', '박쥐', '장어']
#len() = 리스트의 요소 개수를 반환
#print(len(animals))
#인덱스 => 리스트명[i] i = i+1번째 요소에 접근
print(animals[4])
animals[2] = '암소'
print(animals)
#append() => 리스트 가장 마지막에 요소추가
animals.append('여우')
print(len(animals))
#pop() => 삭제할 요소의 인덱스를 통해 삭제
animals.pop(0)
print(animals)
#remove() => 요소의 값을 통해 삭제
animals.remove('장어')
print(animals)
```

```python
#5개의 요소로 이루어진 리스트 만들기
animals = ['개', '개미', '고양이', '박쥐', '장어']
# in => 리스트에 요소가 있으면 True 없으면 False
value = input("확인하고싶은 동물을 입력: ")
# 리스트에 염소가 있는지 확인
if value in animals:
    #index() => 요소의 인덱스를 반환
    print("리스트에", value, "가 있습니다", animals.index(value))
else:
    print("리스트에", value, "가 없습니다")
```

```python
#리스트 정렬
#sort() => 리스트요소를 오름차순으로 정렬
animals =['곰', '여우', '토끼', '사슴', '독수리', '매', '올빼미', '펭귄', '악어',
          '뱀', '거북이', '돌고래', '고래', '상어', '문어', '오징어', '새우',
          '사자', '호랑이', '코끼리', '원숭이', '기린', '얼룩말', '캥거루', '늑대']
animals.sort(reverse=True)
print(animals)
```

```python
#5개의 요소로 이루어진 리스트 만들기
animals = ['개', '개미', '고양이', '박쥐', '장어']

#for 반복문 => 반복횟수를 알고 있을때
for i in animals:
    print(i)
```

```python
#range() => 0~n-1까지 요소를 반환
for i in range(1,1000):
    print(i)
```

```python
#구구단 출력
#2~9
for i in range(2,10):
    for j in range(1,10):
        print(i, "X", j, "=", i*j, end="\t")
    print()
```
