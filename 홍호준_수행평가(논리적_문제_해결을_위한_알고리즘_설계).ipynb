# **캐릭터 초기 스텟 자동 분배기**

초기 포인트를 소진할 때까지 스탯을 분배하며 직업별 필요 최소 능력치
조건을 검사합니다. 게임을 처음 하는 유저들은 종종 자신이 가진 초기
포인트가 10인데도, 힘에만 100을 배분하려 드는 욕심을 부립니다. 게임
서버는 유저의 이런 무리한 요구를 있는 그대로 받아들일 만큼
호락호락하지 않습니다.


• 총 포인트: 초기 50 부여

• 직업 조건: 전사(힘 최소 20), 마법사(지능 최소 20)

• 종료 조건: 남은 포인트가 0이 될 때까지 분배

```python
#총 포인트 설정
total_point = 50
#여러 스텟 변수들 설정
HP=0
ATK=0
INT=0
total_HP=0
total_ATK=0
total_INT=0
status=[]

while True:
    직업=input('원하시는 직업을 선택해 주세요(전사,마법사): ')
    if 직업 == "전사":
        print("전사 최소 조건은 힘 20 입니다")
    elif 직업 == "마법사":
        print("마법사 최소 조건은 지능 20 입니다")
    else:
        print("올바른 직업을 선택해주세요")
        continue

    print(f'가지고 있는 총 포인트는 {total_point}입니다.')

    while True:
        HP=int(input('체력스텟을 찍어주세요: '))
        total_point -= HP
        total_HP += HP
        if total_point < 0:
            print('보유 포인트가 부족합니다.')
            total_point = 50
            HP = 0
            continue
        elif total_point > 0:
            print(f'남은 포인트는 {total_point}입니다')
        else:
            break

        ATK=int(input('힘스텟을 찍어주세요: '))
        total_point -= ATK
        total_ATK += ATK
        if total_point < 0:
            print('보유 포인트가 부족합니다.')
            total_point = 50
            HP = 0
            ATK = 0
            continue
        elif total_point > 0:
            print(f'남은 포인트는 {total_point}입니다')
        else:
            break


        INT=int(input('지능스텟을 찍어주세요: '))
        total_point -= INT
        total_INT += INT
        if total_point < 0:
            print('보유 포인트가 부족합니다.')
            total_point = 50
            HP = 0
            ATK = 0
            INT = 0
            continue
        elif total_point > 0:
            print(f'남은 포인트는 {total_point}입니다')
            continue
        else:
            break

    if 직업 == "전사":
        if total_ATK >= 20:
            break
        else:
            print('직업 최소조건을 맞추지 못했습니다.')
            total_point = 50
            continue

    if 직업 == "마법사":
        if total_INT >= 20:
            break
        else:
            print('직업 최소조건을 맞추지 못했습니다.')
            total_point = 50
            continue

status.insert(0,HP)
status.insert(1,ATK)
status.insert(2,INT)
print(f"분배하신 스텟 체력, 힘, 지능은 각각 {status}입니다")
print('캐릭터가 생성되었습니다.')
```
