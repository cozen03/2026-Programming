#주사위를 굴려서 특정 숫자가 나오면 당첨을 출력하고 아닐경우 낙첨을 출력하는 프로그램을 작성해보자

```python
import random

num = random.randrange(1, 7)

if num == 3:
  print('숫자:', num, '당첨')
else:
  print('숫자:', num, '낙첨')
```
