<font size=5>회전 초밥집에서는 담긴 접시의 색에 따라 초밥의 가격이 다르게 계산된다.<br>다음을 참고하여 회전 초밥의 비용을 영수증 형식으로 출력하는 프로그램을 완성해 보자.<br>

<br><br><br>

##입력예시


```
빨간색 접시의 개수를 입력하세요: 2
노란색 접시의 개수를 입력하세요: 3
파란색 접시의 개수를 입력하세요: 1
초록색 접시의 개수를 입력하세요: 2
```


##출력예시
```
========= 회전 초밥 영수증 =======
 접시     단가     개수       합계
빨간색   1,000       2       2,000
노란색   2,000       3       6,000
파란색   3,000       1       3,000
초록색   4,000       2       8,000
총 가격                     19,000
==================================
```



입력 = 4개
변수 = n개



````python
a = input('빨간색 접시의 개수를 입력하세요: ')
b = input('노란색 접시의 개수를 입력하세요: ')
c = input('파란색 접시의 개수를 입력하세요: ')
d = input('초록색 접시의 개수를 입력하세요: ')
print('========= 회전 초밥 영수증 =======')
print(' 접시\t 단가\t 개수\t합계')
print('빨간색\t1,000\t',a,'\t',int(a)*1000)
print('빨간색\t2,000\t',b,'\t',int(b)*2000)
print('빨간색\t3,000\t',c,'\t',int(c)*3000)
print('빨간색\t4,000\t',d,'\t',int(d)*4000)
print(f'총가격\t\t\t{int(a)*1000+int(b)*2000+int(c)*3000+int(d)*4000}')
print('==================================')
````

````python
red = int(input('빨간색 접시의 개수를 입력하세요: '))
red_price = 1000
red_price_total = red_price * red
sushi_price_total = red_price_total
print('=========회전초밥 영수증==========')
print('접시\t단가\t개수\t합계')
print(f'빨간색\t{red_price}\t{red}\t{red_price_total}')
print(f'총가격\t\t\t{sushi_price_total}')
````

#문제2
<font size=5>적금은 대표적인 금융 상품으로 일정 기간 금액을 납부한 후 이자와 함께 돌려받는 것을 말한다. 다음의 실행 결과를 보고 ‘적금 계산기’ 프로그램을 만들어 보자.
##입력예시

```
월 납입액을 입력하세요 (원): 100000
연 이자율을 입력하세요 (%): 3.75
적금 기간을 입력하세요 (월): 24
```

##출력예시
```
적금 기간 납입한 총 원금: 2400000 원
적금 기간 발생한 총 이자: 7500 원
적금 만기 시 받을 총금액: 2407500 원
```

원금X연이율X기간(년)

````python
월납입액 = float(input('월 납입액을 입력하세요 (원): '))
이자율 = float(input('연 이자율을 입력하세요 (%): '))
기간 = float(input('적금 기간을 입력하세요 (월): '))
총원금 = 월납입액*기간
총이자 = 월납입액*(이자율/100)*기간
총금액 = 총원금 + 총이자
print(f'적금 기간 납입한 총 원금: {총원금}')
print(f'적금 기간 발생한 총 이자: {총이자}')
print(f'적금 만기 시 받을 총금액: {총금액}')
````
