# 변수명 작성규칙

나머지
문자는 영문자, 한글, 밑줄( _ ) , 숫자이어야 한다.


if, for와 같은 예약어(특정한 기능으로 사용되도록 미리 약속된 단어)는 사용할 수 없다.


공백과
하이픈(-)은 사용할 수 없다.


영문자
대소문자를 구별한다.


저장된
데이터를 알 수 있도록 의미 있는 이름을 사용하는 것이 좋다.


여러
단어로 구성된 변수명을 사용하는 경우에는 각 단어를 밑줄( _ ) 로
연결하거나 이어지는 단어를 대문자로 시작하여 가독성을 높이는 것이 좋다(예: sum_result, my_age, sumResult, myAge).

```python
#불 자료형
참 = True
거짓 = False
print(참, type(참))
print(거짓, type(거짓))
print(100 == 20)
print(10 > 5)
```

```python
#문자열 자료형

a = '100'
b = 'python'

#문자열 연산

print(a+b)#문자열과 문자열을 연결
print(a*10)#특정 횟수만큼 반복
print(a < b)#문자의 길이 비교
print('1' in a)#특정 문자열 포함하는지
print('1' not in a)#특정 문자열 포함하지 않는지
```

```python
#문자열 인덱싱과 슬라이싱
a = "Hello"
print(f"첫번째 문자 = {a[0]}")
print(f"마지막 문자 = {a[4]}")
print(f"마지막 문자 = {a[-1]}")
```
