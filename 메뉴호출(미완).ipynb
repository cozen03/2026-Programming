```python
from google.colab import userdata
import requests
import json
import re

def get_menu(date):
    KEY = userdata.get('NEIS_API')
    #기본주소
    url = "https://open.neis.go.kr/hub/mealServiceDietInfo"

    #api 요청 파라미터
    params = {
        "KEY" : KEY,
        "Type" : "json",
        "ATPT_OFCDC_SC_CODE" : "T10",
        "SD_SCHUL_CODE" : "9290056",
        "MMEAL_SC_CODE" : "2", #식사코드 2 = 중식, 3 = 석식
        "MLSV_YMD" : date
    }

    response = requests.get(url,params=params)

    if response.status_code == 200:
        data = response.json()
        menu = data['mealServiceDietInfo'][1]['row'][0]['DDISH_NM']
        menu = menu.replace("<br/>", ',')
        menu = re.sub(r'\([\d\.]+\)', '', menu).replace("  ", " ").strip()
    return menu
```
