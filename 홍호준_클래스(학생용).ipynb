## 1. 클래스는 왜 필요한가?

C 언어와 같은 일부 프로그래밍 언어에는 클래스가 없습니다. 즉, 클래스 없이도 프로그램을 만들 수는 있습니다. 하지만 클래스를 적재적소에 사용하면 프로그래머가 얻을 수 있는 이익이 많습니다.

### 계산기 프로그램을 만들며 클래스 알아보기

계산기는 이전에 계산한 결괏값을 항상 메모리에 저장하고 있어야 합니다. 예를 들어 `3 + 4`의 결괏값 `7`을 기억했다가, 다음 연산 `+ 3`을 수행하면 `7`에 `3`을 더해 `10`을 보여줍니다.

이 '더하기' 기능을 함수로 구현해 보겠습니다.

````python
# calculator.py
result = 0

def add(num):
  # 전역변수 result 선언
  global result
  # 결괏값(result)에 입력값(num) 더하기
  result += num
  # 결괏값 리턴
  return result


print(add(3))
print(add(4))
````

**예상 출력:**
```
3
7
```
이전 계산 결괏값을 유지하기 위해 `result`라는 전역 변수(global)를 사용했습니다.

---
### 문제 상황: 계산기 2대가 필요하다면?

만약 한 프로그램에서 2대의 계산기가 필요하다면 어떻게 해야 할까요? 각 계산기는 독립적인 결괏값을 유지해야 합니다. 위와 같은 `add` 함수 하나만으로는 이를 구현할 수 없습니다.

함수만 사용한다면, 각 계산기를 위한 함수와 전역 변수를 따로 만들어야 합니다.

````python
# calculator2.py

result1 = 0
result2 = 0

def add1(num):  # 계산기1
  # 전역변수 result 선언
  global result1
  # 결괏값(result)에 입력값(num) 더하기
  result1 += num
  # 결괏값 리턴
  return result1
def add2(num):  # 계산기2
  # 전역변수 result 선언
  global result2
  # 결괏값(result)에 입력값(num) 더하기
  result2 += num
  # 결괏값 리턴
  return result2

print(add1(3))
print(add1(4))
print(add2(3))
print(add2(7))
````

**예상 출력:**
```
3
7
3
10
```
계산기 1의 결괏값이 계산기 2에 영향을 끼치지 않습니다.
하지만 계산기가 10개, 100개가 필요하다면? 여기에 '빼기', '곱하기' 기능까지 추가해야 한다면 상황은 매우 복잡해질 것입니다.

---
### 클래스를 사용한 해결

클래스를 사용하면 이 문제를 간단하게 해결할 수 있습니다.

````python
# calculator3.py

class Calculator:
    def __init__(self):
        self.result = 0

    def add(self, num):
        self.result += num
        return self.result

cal1 = Calculator()
cal2 = Calculator()

print(cal1.add(3))
print(cal1.add(4))
print(cal2.add(3))
print(cal2.add(7))
print(cal1.add(10))
print(cal2.add(20))
````

**예상 출력:**
```
3
7
3
10
```
`Calculator` 클래스로 만든 `cal1`과 `cal2`라는 **객체**는 각각의 결괏값을 독립적으로 유지합니다. 계산기 대수가 늘어나도 객체를 생성하기만 하면 되므로 훨씬 간단합니다.

여기에 '빼기' 기능을 추가하는 것도 간단합니다.

````python
class Calculator:
    def __init__(self):
        self.result = 0

    def add(self, num):
        self.result += num
        return self.result

    def sub(self, num):
        self.result -= num
        return self.result

# 테스트
my_cal = Calculator()
print(my_cal.add(10))
print(my_cal.sub(3))
````

## 3. 사칙 연산 클래스 만들기

### 클래스 구상하기

클래스를 만들기 전, 어떻게 동작하게 할지 미리 구상하는 것이 좋습니다.

1.  `a = FourCal()` : `FourCal` 클래스로 `a` 객체를 만듭니다.
2.  `a.setdata(4, 2)` : `a` 객체에 숫자 4와 2를 지정합니다.
3.  `a.add()` : 4 + 2 = 6을 리턴합니다.
4.  `a.mul()` : 4 * 2 = 8을 리턴합니다.
5.  `a.sub()` : 4 - 2 = 2를 리턴합니다.
6.  `a.div()` : 4 / 2 = 2를 리턴합니다.

### 1단계: 클래스 구조 만들기

````python
class FourCal:
    pass

a = FourCal()
type(a)
````

`pass`는 아무것도 수행하지 않는 문법입니다.
`type(a)`의 결과를 보면 `a` 객체가 `FourCal` 클래스의 인스턴스임을 알 수 있습니다.

### 2단계: 객체에 숫자 지정하기 (setdata 메서드)

클래스 안에 구현된 함수를 **메서드(Method)**라고 부릅니다.
사칙연산을 하려면 먼저 두 숫자를 객체에 저장해야 합니다. `setdata`라는 메서드를 만들어 보겠습니다.

````python
class FourCal:
    def setdata(self, first, second):
        self.first = first
        self.second = second
        print(f'{self.first} {self.second}')

a = FourCal()
a.setdata(4, 20)
````

#### 💡 `self`의 정체

`setdata` 메서드는 `self`, `first`, `second` 3개의 매개변수를 받는데,
왜 호출할 때는 `a.setdata(4, 2)`처럼 2개의 값만 전달했을까요?

**이유: 메서드의 첫 번째 매개변수 `self`에는 메서드를 호출한 객체 `a`가 자동으로 전달되기 때문입니다.**

즉, 파이썬이 내부적으로 `a.setdata(4, 2)`를 `FourCal.setdata(a, 4, 2)`처럼 변환해 주는 것입니다.

따라서 `setdata` 메서드 내부의 코드는 다음과 같이 동작합니다.
1. `self.first = first`  -> `a.first = 4`
2. `self.second = second` -> `a.second = 2`

이렇게 `a.first`, `a.second`처럼 객체에 생성되는 변수를 **객체변수** (또는 인스턴스 변수)라고 부릅니다.
---
<img width=1000, src="https://wikidocs.net/images/page/28/05_1_class2.png">

---


````python
a = FourCal()
b = FourCal()

a.setdata(4, 2)
b.setdata(3, 7)

print("a.first =", a.first)  # a 객체의 first 값
print("b.first =", b.first)  # b 객체의 first 값

# b의 값을 바꿨지만 a의 값은 변하지 않습니다.
print("a.first (다시 확인) =", a.first)
````

#### 💡 객체변수는 독립적이다

클래스로 만든 객체들은 각자의 객체변수 값을 독립적으로 유지합니다.

<img width=500 src="https://wikidocs.net/images/page/28/05_1_cookie2.png">

`a` 객체의 `first` 값(4)은 `b` 객체의 `first` 값(3)에 영향을 받지 않고 원래 값을 유지합니다.

### 3단계: 더하기 기능 만들기 (add 메서드)

````python
class FourCal:
    def setdata(self, first, second):
        self.first = first
        self.second = second
        print(f'{self.first} {self.second}')

    def add(self):
        result = self.first + self.second
        return result


a = FourCal()
a.setdata(4, 2)
print(a.add())
````

`a.add()`가 호출되면, `add` 메서드의 `self`는 `a`가 됩니다.
따라서 `result = self.first + self.second` 코드는 `result = a.first + a.second` (즉, `result = 4 + 2`)로 실행됩니다.

### 4단계: 곱하기, 빼기, 나누기 기능 만들기

````python
class FourCal:
    def setdata(self, first, second):
        self.first = first
        self.second = second
        print(f'{self.first} {self.second}')

    def add(self):
        result = self.first+self.second
        return result

    def mul(self):
        result = self.first-self.second
        return result

    def sub(self):
        result = self.first*self.second
        return result

    def div(self):
        result = self.first/self.second
        return result

# 테스트
a = FourCal()
b = FourCal()

a.setdata(4, 2)
b.setdata(3, 8)

print(f"a.add() = {a.add()}")      #
print(f"a.mul() = {a.mul()}")      #
print(f"a.sub() = {a.sub()}")      #
print(f"a.div() = {a.div()}")      #
print(f"b.add() = {b.add()}")      #
print(f"b.mul() = {b.mul()}")      #
print(f"b.sub() = {b.sub()}")      #
print(f"b.div() = {b.div()}")      #
````

## 4. 생성자 (Constructor)

만약 `setdata` 메서드를 호출하는 것을 잊어버리고 `add` 메서드를 호출하면 어떻게 될까요?

````python
try:
    a = FourCal()
    a.add()
except AttributeError as e:
    print(f"오류 발생: {e}")
````

`setdata`를 호출하지 않아 `a.first`와 `a.second` 객체변수가 생성되지 않았기 때문에 `AttributeError`가 발생합니다.

이처럼 객체를 만들 때 **반드시** 설정해야 하는 초깃값들은, **생성자(Constructor)**를 통해 설정하는 것이 안전합니다.

**생성자**는 객체가 생성될 때 (즉, `FourCal()`가 호출될 때) **자동으로 호출되는 메서드**입니다.
파이썬에서는 메서드 이름을 `__init__` (밑줄 두 개)로 만들면 생성자가 됩니다.

`setdata`를 `__init__`으로 바꿔보겠습니다.

````python
class FourCal:
    def __init__(self, first, second):
        print("객체가 생성되었습니다! 숫자를 설정합니다.")
        self.first = first
        self.second = second

    def setdata(self, first, second):
        self.first = first
        self.second = second

    def add(self):
        return self.first + self.second

    def mul(self):
        return self.first * self.second

    def sub(self):
        # ※ 원문 오타 수정
        return self.first - self.second

    def div(self):
        return self.first / self.second
````

이제 `__init__` 생성자는 `first`와 `second` 두 개의 인수를 필요로 합니다.
따라서 `a = FourCal()`처럼 인수를 전달하지 않고 객체를 생성하려고 하면 오류가 발생합니다.

````python
try:
    a = FourCal(10,6)
except TypeError as e:
    print(f"오류 발생: {e}")
````

객체를 생성할 때 `__init__` 메서드가 요구하는 인수를 **반드시** 전달해야 합니다.

````python
# 객체를 생성하는 시점에 (4, 2) 값을 전달합니다.
a = FourCal(4, 2)

# __init__ 메서드가 자동으로 호출되어 객체변수가 생성되었습니다.
print(f"a.first = {a.first}")
print(f"a.second = {a.second}")

# 이제 setdata를 따로 호출할 필요 없이 바로 메서드를 사용할 수 있습니다.
print(f"a.add() = {a.add()}")
print(f"a.div() = {a.div()}")
````

## 5. 클래스의 상속 (Inheritance)

**상속**은 '물려받다'라는 뜻으로, 어떤 클래스가 다른 클래스의 기능을 그대로 물려받는 것입니다.

상속은 **기존 클래스는 그대로 둔 채, 새로운 기능을 추가**하거나 **기존 기능을 변경**하고 싶을 때 사용합니다.

예를 들어, `FourCal` 클래스의 기능(더하기, 빼기 등)은 그대로 유지하면서, $a^b$ (거듭제곱) 기능을 추가해 보겠습니다.

상속 문법: `class 새로운_클래스(상속할_클래스):`


```
상속 기능은 왜 쓰는 걸까?
```
보통 상속은 기존 클래스를 변경하지 않고 기능을 추가하거나 기존 기능을 변경하려고 할 때 사용한다. ‘클래스에 기능을 추가하고 싶으면 기존 클래스를 수정하면 되는데 왜 굳이 상속을 받아서 처리해야 하지?’라는 의문이 들 수도 있다. 하지만 기존 클래스가 라이브러리 형태로 제공되거나 수정이 허용되지 않는 상황이라면 상속을 사용해야 한다.



````python
# 1. 'FourCal' 클래스는 이전 셀(Cell 15)에서 정의된 버전을 사용합니다.

# 2. FourCal 클래스를 상속하는 MoreFourCal 클래스를 만듭니다.
# ※ 원문의 'More FourCal'는 오타입니다. 'MoreFourCal'로 수정했습니다.
class MoreFourCal(FourCal):
    pass # 우선 아무 기능도 추가하지 않습니다.

# 3. MoreFourCal 객체를 생성합니다.
# 부모(FourCal)의 __init__ 생성자가 호출되므로 (4, 2)를 전달해야 합니다.
a = MoreFourCal(4, 2)

# 4. 물려받은 기능을 사용해 봅니다.
print(f"a.add() = {a.add()}")
print(f"a.mul() = {a.mul()}")
````

`MoreFourCal`은 `FourCal`의 모든 기능을 물려받았습니다.
이제 여기에 새로운 기능인 `pow` (거듭제곱) 메서드를 추가해 보겠습니다.

````python
# ※ 원문의 'More FourCal'는 오타입니다. 'MoreFourCal'로 수정했습니다.
class MoreFourCal(FourCal):
    def pow(self):
        result = self.first ** self.second  # ** 기호는 거듭제곱 연산자입니다.
        return result

a = MoreFourCal(4, 2)

# 새로 추가한 pow 메서드 테스트
print(f"a.pow() = {a.pow()}")

# 물려받은 add 메서드 테스트
print(f"a.add() = {a.add()}")
````

## 6. 메서드 오버라이딩 (Method Overriding)

만약 부모 클래스(`FourCal`)의 `div` 메서드를 0으로 나눌 때 오류가 발생하는 것이 아니라, 0을 리턴하도록 바꾸고 싶다면 어떻게 해야 할까요?

부모 클래스에 있는 메서드를 자식 클래스에서 **동일한 이름으로 다시 만드는 것**을 **메서드 오버라이딩**이라고 합니다.

이렇게 하면, 부모의 메서드 대신 자식 클래스가 '덮어쓴(overriding)' 메서드가 호출됩니다.

````python
class SafeFourCal(FourCal):
    # div 메서드를 오버라이딩 (재정의)
    def div(self):
        if self.second == 0:  # 나누는 값이 0인 경우
            return 0  # 0을 리턴
        else:
            # 0이 아니면, 부모(FourCal)의 원래 기능을 그대로 사용
            return self.first / self.second

# 테스트
a = SafeFourCal(4, 0)
print(f"SafeFourCal(4, 0) -> a.div() = {a.div()}") # 오류 대신 0이 출력됩니다.

b = SafeFourCal(4, 2)
print(f"SafeFourCal(4, 2) -> b.div() = {b.div()}") # 0이 아닐 땐 정상적으로 계산됩니다.
````

## 7. 클래스변수 (Class Variable)

지금까지 배운 `self.first` 같은 변수는 **객체변수**입니다. 객체마다 독립적인 값을 가집니다.

**클래스변수**는 클래스 안에 선언된 변수로, 그 클래스로 만든 **모든 객체가 공유**하는 변수입니다.

````python
class Family:
    lastname = "김"  # 클래스변수 선언

# 클래스 이름으로 직접 접근 가능
print(f"Family.lastname = {Family.lastname}")

# 객체를 생성
a = Family()
b = Family()

# 객체를 통해서도 접근 가능
print(f"a.lastname = {a.lastname}")
print(f"b.lastname = {b.lastname}")

print("--- 클래스변수 변경 ---")
Family.lastname = "박"

# 클래스변수를 변경하면 모든 객체에 반영됩니다.
print(f"a.lastname = {a.lastname}")
print(f"b.lastname = {b.lastname}")
````

### ※ 클래스변수와 동일한 이름의 객체변수

만약 `a.lastname = "최"` 처럼 객체를 통해 값을 변경하면 어떻게 될까요?

이 경우, 클래스변수 `Family.lastname`이 바뀌는 것이 아니라, `a` 객체에 `lastname`이라는 **새로운 객체변수**가 생성됩니다.
이 객체변수는 클래스변수와 이름이 같지만, 완전히 다른 변수입니다.

````python
print(f"현재 Family.lastname = {Family.lastname}")
print(f"현재 a.lastname = {a.lastname}")
print(f"현재 b.lastname = {b.lastname}")

print("--- 'a' 객체에만 객체변수 할당 ---")
a.lastname = "최"

print(f"a.lastname = {a.lastname}")  # 'a'는 새로 생성된 객체변수 '최'를 참조
print(f"b.lastname = {b.lastname}")  # 'b'는 여전히 클래스변수 '박'을 참조
print(f"Family.lastname = {Family.lastname}") # 클래스변수는 '박' 그대로임
````

#👩‍💻 과제: Student 클래스 만들기
목표: 학생 정보를 관리하는 Student 클래스를 정의합니다. 이 클래스는 학생 개개인의 고유 정보(이름, 전공, 이수 학점)와 모든 학생이 공유하는 정보(총 학생 수, 졸업 기준 학점)를 모두 다룰 수 있어야 합니다.

📜 요구 사항

1.**클래스 변수 (모든 학생이 공유)**

Student 클래스에는 다음과 같은 클래스 변수 2개가 필요합니다.
*   total_students: 총 학생 수를 기록합니다. (초기값: 0)
*   min_credits_for_graduation: 졸업에 필요한 최소 이수 학점을 정의합니다. (초기값: 130)




2.**객체 변수 (학생별 고유 정보)**

__init__ 생성자 메소드를 사용하여, 학생 객체가 생성될 때 다음 객체 변수들을 초기화해야 합니다.


*   name: 이름 (문자열)
*   major: 전공 (문자열)
*   grade: 학년 (숫자)
*   credits_taken: 현재 이수 학점 (반드시 0으로 시작)


**[중요] __init__ 메소드 안에서 객체가 생성될 때마다, 클래스 변수 total_students의 값을 1씩 증가시켜야 합니다.**

3.메소드 (학생의 기능)

다음 기능을 수행하는 3개의 메소드를 정의해야 합니다.

<font size=5 color='blue'>introduce(self)</font>
```

학생의 정보를 출력합니다.
(예시 출력: "안녕하세요, 저는 [전공] [학년]학년 [이름]입니다.")
```

<font size=5 color='blue'>take_class(self, credits)</font>
```

학생이 수업을 들어 학점을 이수하는 기능입니다.

credits (숫자) 값을 매개변수로 받습니다.

학생의 객체 변수 credits_taken에 credits만큼 더해줍니다.

(예시 출력: "[이름]" 학생이 [credits]학점을 이수했습니다. (총 이수 학점: [credits_taken])")
```


<font size=5 color='blue'>check_graduation(self)</font>
```
학생의 졸업 가능 여부를 확인합니다.

객체 변수 self.credits_taken과 클래스 변수 Student.min_credits_for_graduation을 비교합니다.

(조건 1) 이수 학점이 졸업 기준 학점 이상일 경우:

(예시 출력: "[이름]" 학생은 졸업 요건을 충족했습니다! (총 130학점 중 [credits_taken]학점 이수)")

(조건 2) 이수 학점이 졸업 기준 학점 미만일 경우:

부족한 학점을 계산해야 합니다. (부족한 학점 = 130 - credits_taken)

(예시 출력: "[이름]" 학생은 졸업까지 [부족한 학점]학점이 더 필요합니다.")
```

````python
class Student:
    total_students = 0
    min_credits_for_graduation = 0
    def __init__(self,name,major,grade):
        self.name = name
        self.major = major
        self.grade = grade
        self.credits_taken = 0
        Student.total_students += 1
        Student.min_credits_for_graduation += 130
    def take_class(self,credits):
        self.credits_taken += credits
        print(f"{self.name}학생이 {credits}학점을 이수했습니다. (총 이수 학점: {self.credits_taken})")
    def introduce(self):
        print(f"안녕하세요, 저는 {self.major} {self.grade}학년 {self.name}입니다.")
    def check_graduation(self):
        if self.credits_taken >= Student.min_credits_for_graduation:
            print(f"{self.name} 학생은 졸업 요건을 충족했습니다! (총 130학점 중 {self.credits_taken}학점 이수)")
        else:
            print(f"{self.name} 학생은 졸업까지 {Student.min_credits_for_graduation-self.credits_taken}학점이 더 필요합니다.")
````

````python
str1 = Student("홍호준","컴퓨터공학과",2)
str1.introduce()
str1.take_class(55)
str1.take_class(55)
str1.take_class(20)
str1.check_graduation()
````

````python
str1 = Student("홍호준","컴퓨터공학과",2)
str1.introduce()
str1.take_class(55)
str1.take_class(20)
str1.check_graduation()
````
