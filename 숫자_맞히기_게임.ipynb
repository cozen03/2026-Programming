```python
#upper() => 문자열 전체를 대문자로 변환
#strip() => 양쪽 공백 제거
userInput = ""

while userInput != "STOP":
    userInput = input("무언가 입력하거나 입력을 멈추려먼 STOP을 입력하세요.").upper().strip()

```

```python
animals = []

userInput = " "

while userInput != "":
    userInput = input("동물을 입력하세요.(끝내려면 빈칸으로 두세요.)").upper().strip()

    if len(userInput) > 0:
        animals.append(userInput)
print(animals)
```

```python
import random

randNum = random.randrange(1,101)#랜덤범위 부여

userInput = " "
count = 0
while userInput != randNum:
    userInput = input()
    if userInput.isnumeric(): #숫자만 입력되었는지 확인
        userInput = int(userInput)
        if userInput < 0 or userInput > 100: #에외상황 제거
            print("범위를 벗어났습니다.")
            continue
        else:
            count += 1
            if userInput > randNum:
                print("정답이 더 낮아요.")
                print(f"시도횟수{count}")
            elif userInput < randNum:
                print("정답이 더 높아요.")
                print(f"시도횟수{count}")
            else:
                print("정답이에요.")
                print(f"시도횟수{count}")
    else:
        print("잘못된 값이 입력되었습니다.") #예외상황 제거
        continue




```
