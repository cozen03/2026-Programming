```python
import random

# [함수 정의] 스탯 분배 머신
def 스탯_분배_함수(보너스_포인트):
    """
    주어진 포인트 한도 내에서 사용자에게 체력, 힘, 지능, 방어력 스텟을 입력받아 반환하는 함수
    """
    while True:
        남은_포인트_풀 = 보너스_포인트
        print(f'\n분배 가능한 스탯 포인트는 {남은_포인트_풀}입니다.')

        try:
            입력_체력 = int(input(f'체력 스텟을 찍어주세요 (남은 포인트: {남은_포인트_풀}): '))
            if 입력_체력 > 남은_포인트_풀 or 입력_체력 < 0:
                print("[경고] 잘못된 입력이거나 포인트가 부족합니다. 다시 입력하세요.")
                continue
            남은_포인트_풀 -= 입력_체력

            입력_힘 = int(input(f'힘 스텟을 찍어주세요 (남은 포인트: {남은_포인트_풀}): '))
            if 입력_힘 > 남은_포인트_풀 or 입력_힘 < 0:
                print("[경고] 잘못된 입력이거나 포인트가 부족합니다. 다시 입력하세요.")
                continue
            남은_포인트_풀 -= 입력_힘

            입력_지능 = int(input(f'지능 스텟을 찍어주세요 (남은 포인트: {남은_포인트_풀}): '))
            if 입력_지능 > 남은_포인트_풀 or 입력_지능 < 0:
                print("[경고] 잘못된 입력이거나 포인트가 부족합니다. 다시 입력하세요.")
                continue
            남은_포인트_풀 -= 입력_지능

            입력_방어력 = int(input(f'방어력 스텟을 찍어주세요 (남은 포인트: {남은_포인트_풀}): '))
            if 입력_방어력 > 남은_포인트_풀 or 입력_방어력 < 0:
                print("[경고] 잘못된 입력이거나 포인트가 부족합니다. 다시 입력하세요.")
                continue
            남은_포인트_풀 -= 입력_방어력

            if 남은_포인트_풀 != 0:
                print(f"[경고] 포인트를 정확히 {보너스_포인트}만큼 다 사용해야 합니다. (남은 포인트: {남은_포인트_풀})")
                continue

            return 입력_체력, 입력_힘, 입력_지능, 입력_방어력

        except ValueError:
            print("[경고] 숫자만 입력해주세요!")

# 캐릭터 생성 및 직업 선택
print("=================== RPG GAME START ===================")
while True:
    직업 = input('원하시는 직업을 선택해 주세요 (전사, 마법사): ')
    if 직업 in ["전사", "마법사", "도박사"]:
        if 직업 == "전사": print("전사 최소 조건은 힘 20 입니다.")
        else: print("마법사 최소 조건은 지능 20 입니다.")
        break
    print("[경고] 올바른 직업을 선택해주세요.")

while True:
    초기_체력, 초기_힘, 초기_지능, 초기_방어력 = 스탯_분배_함수(100)

    if 직업 == "전사" and 초기_힘 < 20:
        print('[경고] 전사 조건을 만족하지 못했습니다. 힘을 20 이상으로 다시 찍으세요.')
        continue
    elif 직업 == "마법사" and 초기_지능 < 20:
        print('[경고] 마법사 조건을 만족하지 못했습니다. 지능을 20 이상으로 다시 찍으세요.')
        continue
    올스텟 = [초기_체력, 초기_힘, 초기_지능, 초기_방어력]
    if 초기_지능 == 77:
        초기_체력 = 0
        초기_힘 = 0
        초기_지능 = 0
        초기_방어력 = 0
        print("히든 캐릭터 도박사가 해금 되었습니다.")
        print("도박사 최소조건은 올스텟 999입니다.")
        직업 = "도박사"
        while True:
            초기_체력, 초기_힘, 초기_지능, 초기_방어력 = 스탯_분배_함수(999*4)

            if 직업 == "도박사" and 올스텟[0] == 999 or 올스텟[1] == 999 or 올스텟[2] == 999 or 올스텟[3] == 999:
                print('[경고] 도박사 조건을 만족하지 못했습니다. 올스텟을 999 이상으로 다시 찍으세요.')
                continue
            break
    플레이어_레벨 = 1
    누적_체력스텟 = 초기_체력
    누적_힘스텟 = 초기_힘
    누적_지능스텟 = 초기_지능
    누적_방어력스텟 = 초기_방어력

    break

print(f"\n[{직업}] 캐릭터 생성이 완료되었습니다! (레벨: {플레이어_레벨})")


# 5개 스테이지 및 랜덤 몬스터 전투 루프
# 몬스터 도감 리스트 (이름, 기본체력, 기본공격력)
몬스터_도감 = [
    {"이름": "슬라임", "기본체력": 80, "기본공격력": 7},
    {"이름": "고블린", "기본체력": 130, "기본공격력": 12},
    {"이름": "늑대", "기본체력": 150, "기본공격력": 17},
    {"이름": "늑대 우두머리", "기본체력": 200, "기본공격력": 27},
    {"이름": "오우거", "기본체력": 230, "기본공격력": 22},
    {"이름": "켈베로스", "기본체력": 300, "기본공격력": 37},
    {"이름": "와이번", "기본체력": 250, "기본공격력": 42},
]

게임_클리어_여부 = True

# 1스테이지부터 5스테이지까지 반복
for 스테이지 in range(1, 6):
    print("\n" + "-"*35)
    print(f"STAGE {스테이지} 에 진입합니다.")
    print("-"*35)

    # 스테이지 진입 시 현재 체력 리셋 (스텟의 10배 적용)
    플레이어_최대체력 = 누적_체력스텟 * 10
    플레이어_현재체력 = 플레이어_최대체력

    print(f"체력이 회복되었습니다. (최대 HP: {플레이어_최대체력})")

    # 몬스터 풀에서 랜덤으로 한 마리 선택
    선택된_몬스터 = random.choice(몬스터_도감)
    몬스터_이름 = 선택된_몬스터["이름"]

    # 후반 스테이지 보정
    몬스터_최대체력 = 선택된_몬스터["기본체력"] + (스테이지 * 20)
    몬스터_현재체력 = 몬스터_최대체력
    몬스터_공격력 = 선택된_몬스터["기본공격력"] + (스테이지 * 8)

    print(f"{몬스터_이름}이(가) 등장했습니다! (체력: {몬스터_현재체력}, 공격력: {몬스터_공격력})")

    # 1:1 턴제 전투 루프
    턴_번호 = 1
    while 플레이어_현재체력 > 0 and 몬스터_현재체력 > 0:
        print(f"\n--- [Turn {턴_번호}] ---")
        print(f"내 상태 - 체력: {플레이어_현재체력}/{플레이어_최대체력} 방어력: {누적_방어력스텟}")
        print(f"{몬스터_이름} 상태 - 체력: {몬스터_현재체력}/{몬스터_최대체력}")

        방어_상태 = False # 턴 시작 시 방어 상태 초기화
        행동_선택 = input("행동을 선택하세요 (1: 공격, 2: 방어, 3: 도망): ")

        # [플레이어 공격 턴]
        if 행동_선택 == "1":
            if 직업 == "전사":
                가한_데미지 = random.randint(누적_힘스텟 - 3, 누적_힘스텟 + 5)
                print(f"전사의 베기 공격! {몬스터_이름}에게 {가한_데미지}의 데미지!")
            elif 직업 == "마법사":
                가한_데미지 = random.randint(누적_지능스텟 - 2, 누적_지능스텟 + 7)
                print(f"마법사의 파이어볼 공격! {몬스터_이름}에게 {가한_데미지}의 데미지!")
            else:
                총_가한_데미지 = 0
                가한_데미지 = random.randint(누적_힘스텟 - 3, 누적_힘스텟 + 5)
                총_가한_데미지 += 가한_데미지
                가한_데미지 = random.randint(누적_지능스텟 - 2, 누적_지능스텟 + 7)
                총_가한_데미지 += 가한_데미지
                print(f"도박사의 코인 발동! {몬스터_이름}에게 {총_가한_데미지}의 데미지!")
            몬스터_현재체력 -= 가한_데미지
            if 몬스터_현재체력 <= 0:
                몬스터_현재체력 = 0
                break

        # [플레이어 방어 턴]
        elif 행동_선택 == "2":
            방어_상태 = True
            print("이번 턴에 집중적으로 방어합니다!")

        elif 행동_선택 == "3":
            print("도망쳤습니다. 게임 오버.")
            플레이어_현재체력 = 0
            break
        else:
            print("[경고] 잘못된 입력으로 행동 기회를 날렸습니다!")

        # [몬스터 공격 턴]
        if 몬스터_현재체력 > 0:
            몬스터_기본_데미지 = random.randint(몬스터_공격력 - 2, 몬스터_공격력 + 2)

            # 기본 방어 계산
            받은_데미지 = 몬스터_기본_데미지 - 누적_방어력스텟

            # 방어 행동 시 추가 방어력만큼 또 감소
            if 방어_상태:
                받은_데미지 -= 누적_방어력스텟
                print(f"방어 성공! 추가로 {누적_방어력스텟}의 데미지를 더 상쇄했습니다.")

            # 최종 데미지 1 보장
            if 받은_데미지 < 1:
                받은_데미지 = 1

            print(f"{몬스터_이름}의 공격! 최종 {받은_데미지}의 데미지를 입었습니다.")

            플레이어_현재체력 -= 받은_데미지
            if 플레이어_현재체력 <= 0:
                플레이어_현재체력 = 0

        턴_번호 += 1

    # 전투 결과 처리
    if 플레이어_현재체력 <= 0:
        print(f"\n패배했습니다. {몬스터_이름}에게 무릎을 꿇었습니다.")
        게임_클리어_여부 = False
        break
    else:
        print(f"\n승리! {몬스터_이름}을(를) 처치하고 스테이지 {스테이지}를 클리어했습니다.")
        스테이지_보너스_스탯 = (스테이지//2)*15
        if 스테이지_보너스_스탯 < 1:
            스테이지_보너스_스탯 += 15

        # 보스전이 아니라면 레벨업 및 보너스 스탯 분배 진행
        if 직업 == "도박사":
            print("더 이상의 레벨업은 필요없어 보입니다.....")
        elif 스테이지 < 6:
            플레이어_레벨 += 1
            print(f"\nLEVEL UP! 레벨 {플레이어_레벨}이 되었습니다!")
            print(f"레벨업 보너스 스탯 포인트 {스테이지_보너스_스탯}개가 주어집니다.")

            # 함수를 재사용해 보너스 스탯 15개를 새로 분배받음
            보너스_체력, 보너스_힘, 보너스_지능, 보너스_방어력 = 스탯_분배_함수(스테이지_보너스_스탯)

            # 기존 누적 능력치에 추가 합산
            누적_체력스텟 += 보너스_체력
            누적_힘스텟 += 보너스_힘
            누적_지능스텟 += 보너스_지능
            누적_방어력스텟 += 보너스_방어력
            print(f"현재 누적 스텟 -> 체력: {누적_체력스텟}, 힘: {누적_힘스텟}, 지능: {누적_지능스텟}, 방어력: {누적_방어력스텟}")

# [3단계] 최종 보스 스테이지 (드래곤 vs 네크로맨서 랜덤 등장)
보스_도감 = [
    {"이름": "화염의 고룡", "기본체력": 1200, "기본공격력": 80},
    {"이름": "영혼 네크로맨서", "기본체력": 1000, "기본공격력": 75}
]

if 게임_클리어_여부:
    print("\n" + "="*40)
    print("경고: 거대한 기운이 느껴집니다. 최종 보스가 나타났습니다!")
    print("="*40)

    # 마지막 전투 전 체력/마나 최종 풀 리셋
    플레이어_최대체력 = 누적_체력스텟 * 10
    플레이어_현재체력 = 플레이어_최대체력
    print(f"보스전을 위해 모든 체력과 마나가 충전되었습니다!")

    # 보스 랜덤 선택
    if 직업 == "도박사":
        보스_이름 = "ss급 좀비드래곤"
        보스_최대체력 = 99999
        보스_현재체력 = 보스_최대체력
        보스_공격력 = 9999
        대답 = input("도박사의코인 스킬을 발동 하시겠습니까?(네/아니요): ")
        if 대답 == "네":
            print("도박사의 코인 스킬이 발동되었습니다!")
            누적_체력스텟 += 9000
            누적_힘스텟 += 9000
            누적_지능스텟 += 9000
            누적_방어력스텟 += 9000
    else:
        선택된_보스 = random.choice(보스_도감)
        보스_이름 = 선택된_보스["이름"]
        보스_최대체력 = 선택된_보스["기본체력"]
        보스_현재체력 = 보스_최대체력
        보스_공격력 = 선택된_보스["기본공격력"]

    print(f"\n최종 보스 [{보스_이름}]이(가) 등장했습니다! (체력: {보스_최대체력}, 공격력: {보스_공격력})")

    보스_턴 = 1
    while 플레이어_현재체력 > 0 and 보스_현재체력 > 0:
        print(f"\n--- [보스전 Turn {보스_턴}] ---")
        print(f"내 상태 - 체력: {플레이어_현재체력}/{플레이어_최대체력} 방어력: {누적_방어력스텟}")
        print(f"[{보스_이름}] 상태 - 체력: {보스_현재체력}/{보스_최대체력}")

        방어_상태 = False
        행동_선택 = input("행동을 선택하세요 (1: 공격, 2: 방어, 3: 도망): ")

        if 행동_선택 == "1":
            if 직업 == "전사":
                가한_데미지 = random.randint(누적_힘스텟 - 3, 누적_힘스텟 + 5)
                print(f"전사의 베기 공격! {보스_이름}에게 {가한_데미지}의 데미지!")
            elif 직업 == "마법사":
                가한_데미지 = random.randint(누적_지능스텟 - 2, 누적_지능스텟 + 7)
                print(f"마법사의 파이어볼 공격! {보스_이름}에게 {가한_데미지}의 데미지!")
            else:
                총_가한_데미지 = 0
                가한_데미지 = random.randint(누적_힘스텟 - 3, 누적_힘스텟 + 5)
                총_가한_데미지 += 가한_데미지
                가한_데미지 = random.randint(누적_지능스텟 - 2, 누적_지능스텟 + 7)
                총_가한_데미지 += 가한_데미지
                print(f"도박사의 코인 발동! {보스_이름}에게 {총_가한_데미지}의 데미지!")
            보스_현재체력 -= 가한_데미지
            if 보스_현재체력 <= 0:
                보스_현재체력 = 0
                break

        # 보스 공격 턴
        if 보스_현재체력 > 0:
            보스_기본_데미지 = random.randint(보스_공격력 - 5, 보스_공격력 + 5)
            받은_데미지 = 보스_기본_데미지 - 누적_방어력스텟

            if 방어_상태:
                받은_데미지 -= 누적_방어력스텟
                print(f"방어 철벽! 보스의 공격을 받아내어 {누적_방어력스텟}의 데미지를 더 줄였습니다.")

            if 받은_데미지 < 1:
                받은_데미지 = 1

            print(f"[{보스_이름}]의 치명적인 공격! 최종 {받은_데미지}의 데미지를 입었습니다.")
            플레이어_현재체력 -= 받은_데미지
            if 플레이어_현재체력 <= 0:
                플레이어_현재체력 = 0

        보스_턴 += 1

    if 플레이어_현재체력 <= 0:
        게임_클리어_여부 = False
    else:
        print(f"\n위대한 승리! 최종 보스 [{보스_이름}]을(를) 물리쳤습니다!")

# [4단계] 최종 엔딩 평가
print("\n" + "="*40)
if 게임_클리어_여부:
    print("축하합니다! 모든 시련을 이겨내고 세계의 평화를 되찾았습니다!")
else:
    print("GAME OVER.. 보스의 벽은 높았습니다. 다시 도전해 보세요.")
print("="*40)
```
