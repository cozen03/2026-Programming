```python
def op(num1,num2):
  plus = num1 + num2
  minus = num1 - num2
  multiply = num1 * num2
  divide = num1 / num2
  return plus, minus, multiply, divide

a, b = map(int, input('입력: ').split())
p, m, mul, d = op(a,b)
print(f'{a} + {b} = {p}')
print(f'{a} - {b} = {m}')
print(f'{a} * {b} = {mul}')
print(f'{a} / {b} = {d}')
```
